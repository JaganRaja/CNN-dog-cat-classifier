```python
#library to move
import shutil
import imghdr
```

```python
#basic image preprocessing.. if image is corrupted model throws error while training/Prediction
#checking whether the image is in proper format or not
    #whether the image can be openable or not
    #if the file is corrupted move to bad folder

for dirs in os.listdir(PARENT_DIR):
    full_path_data_dirs = os.path.join(PARENT_DIR,dirs)
    
    for imgs in os.listdir(full_path_data_dirs):
        path_to_img = os.path.join(full_path_data_dirs, imgs)
        try:
            img = Image.open(path_to_img)
            img.verify()
            #channel != 3 or image not in jepg/png
            if len(img.getbands()) != 3 or imghr.what(path_to_img) not in ['jpeg','png']:
                bad_data_path = os.path.join(BAD_DATA_DIRECTORY, imgs)
                shutil.move(path_to_img, bad_data_path)
                continue

            print(f"{path_to_img} is verifed with format{imghr.what(path_to_img)}")
        except Exception as e:
            print(f"{path_to_img} is bad")
            #if the file is corrupted move to bad folder
            bad_data_path = os.path.join(BAD_DATA_DIRECTORY, imgs)
            shutil.move(path_to_img, bad_data_path)
                  
```

```python
# creating dataset genertor which helps to load the dataset
#our image dataset can have different image shape and size. we need to keep all the images in a same size and 
# in square shape.. why square shape ?
# 1. W&H calculation willbe easier
# 2. While applying filtering on the image(convolution layer), it gives good result. If the image is rectangle, while filtering
# it may miss the corner values of the image
# keeping the image size depends on memory.. big size occupies more RAM
IMG_SIZE = (180, 180)
BATCH_SIZE = 32
```

```python
#image_dataset_from_directory.. refer docs of this method
    #path
    #validation_split
    #subset
    #seed
    #image_size
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    PARENT_DIR,
    validation_split=0.2,
    subset = "training",
    seed = 42,
    image_size = IMG_SIZE,
    batch_size = BATCH_SIZE  
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    PARENT_DIR,
    validation_split=0.2,
    subset = "validation",
    seed = 42,
    image_size = IMG_SIZE,
    batch_size = BATCH_SIZE  
)
```

```python
# while training some images we keep it as a buffer(keep Images are ready to feed to the training)
train_ds = train_ds.prefetch(buffer_size=32)
test_ds = test_ds.prefetch(buffer_size=32) # maybe error
```

```python
LAYERS = [
    # 1st conv2d layer
    tf.keras.layers.Conv2D(32, (3,3), input_shape=(180,180,3), activation="relu"),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    #2st conv2d layer
    tf.keras.layers.Conv2D(32, (3,3), activation="relu"),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    #3rd conv2d layer
    tf.keras.layers.Conv2D(32, (3,3), activation="relu"),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    # Flattening
    tf.keras.layers.Flatten(),
    #adding dense layer
    tf.keras.layers.Dense(128, activation="relu"),
    #final layer
    tf.keras.layers.Dense(2, activation="softmax") 
]

classifier = tf.keras.Sequential(LAYERS)
```

```python
classifier.compile(
    optimizer=tf.keras.optimizers.Adam(0.01),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```
